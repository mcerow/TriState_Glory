# Tri-State Glory

We're tasked with choosing the best zip codes in the tri-state area to buy in. We made the decision by looking at the return on investement between 1996 and 2018. From there we landed on our 9 zip codes that we'll further explore in subsequent notebooks. This notebook attains the original Zillow data and filters down to the zip codes we want to investigate.

The first step is to import the necessary libraries to get the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import missingno as mno
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', None)
# setting format for readability
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# read in the csv

df = pd.read_csv('zillow_data.csv')
df.head()
```

## Data Preprocessing

The Zillow file reads in as wide format. We'll need to melt it down to long format to do our time series analysis, but that will happen in a separate notebook. The first thing we want to do is change the datatype for all of the observations starting in April 1996. This function has been provided to us by the Flatiron School.

```python
# provided function

def get_datetimes(df):
    return pd.to_datetime(df.columns.values[7:], format='%Y-%m')
```

```python
# pass Zillow data through get_datetimes function

get_datetimes(df)
```

```python
# checking the shape of our data

df.shape
```

We have 272 observed zip codes from 1996 and 2018 across the United States. We'll be focusing on the tristate area which we filter for below.

In order to determine the 'best' zip codes, we want to see where the best return is. We added the ROI column to determine which zip codes we want to model for.

```python
df['ROI'] = (df['2018-04']/df['1996-04'])-1
```

## Filtering for Zipcodes

We're interesting at looking only in the tristate area. We filter our dataframe for just NY, NJ and CT. Once we filter, we see that we've cut our number of zip codes observed from 14.7k to 1.6k. We also want to check for null values.

```python
tristate = df[df['State'].isin(['NY', 'NJ', 'CT'])]
tristate.shape
```

We have null values in the Metro column and the ROI column. We're not concerned with the Metro column in this project, but we do need to understand the null values in the ROI column. When we filtered for those values we saw that there is no recored data for these 30 zip codes from 1996 to 2004. We made the decision to only look at values that have had observations for the full time period and dropped these observations.

```python
tristate.isnull().sum()
```

```python
tristate[tristate['ROI'].isnull()]
```

```python
tristate.dropna(subset=['ROI'],inplace=True)
```

To keep as a reference, we added the 'AveragePrice' column to see the average price over all the time period. This is just for reference to gain more domain knowledge on the data we're working with. This column will dropped once we melt the data into the long form and begin modeling.

```python
tristate['AveragePrice'] = tristate.loc[:,'1996-04':'2018-04'].mean(axis=1)
```

```python
tristate.shape
```

```python
# rename RegionName to Zipcode to be more interpretable

tristate.rename({'RegionName': 'Zipcode'}, axis='columns', inplace=True)
```

We first wanted to check the top 5 zip codes based on the ROI. When we filtered for this, it looks very similiar - lots of Brooklyn only interrupted by Jersey City. We want to offer a more robust look at the tristate area. Instead of only looking at the top 5 zip codes, we decided to look at the top 3 in each state.

```python
tristate.sort_values('ROI',axis=0,ascending=False)[:5]
```

```python
# saving each state as its own dataframe to then filter for

new_york = tristate[tristate['State'] == 'NY']
new_jersey = tristate[tristate['State'] == 'NJ']
connecticut = tristate[tristate['State'] == 'CT']
```

```python
new_york[['Zipcode', 'City', 'State', 'Metro', 'CountyName', 'AveragePrice', 'ROI']].sort_values('ROI',axis=0,ascending=False)[:3]
```

The top 3 zip codes for New York are all in Brooklyn. 11211 is Williamsburg. 11222 is Greenpoint and 11216 covers part of Bedford-Stuyvesant next to Clinton Hill and Stuyvesant Heights.

```python
new_jersey[['Zipcode', 'City', 'State', 'Metro', 'CountyName', 'AveragePrice', 'ROI']].sort_values('ROI',axis=0,ascending=False)[:3]
```

For New Jesery, 07302 Jersey City has the highest ROI, followed by 07030 Hoboken. Behind Hoboken is the Jesery Shore town 07756 Ocean Grove.

```python
connecticut[['Zipcode', 'City', 'State', 'Metro', 'CountyName', 'AveragePrice', 'ROI']].sort_values('ROI',axis=0,ascending=False)[:3]
```

Unlike New York and New Jersey, the zip codes in Connecticut that had the biggest ROI were not cities. The ROI was much more modest in Connecticut compared to the other two states. 06820 is Darien, CT is located right near Stamford along the Long Island Sound. 06068 is Salisbury, CT located in the corner of CT along the Massachusetts and New York border. The final zip code for Connecticut is 06793, Washington, CT also near the New York border.

```python
# save new dataframe for the 9 zip codes we want to model for later

zip_codes = [11211, 11222, 11216, 7302, 7030, 7756, 6820, 6068, 6793]

zipcodes = tristate[tristate['Zipcode'].isin(zip_codes)]
zipcodes.head()
```

The final step is to save the data as a new csv. We'll work off of this csv to melt our data and later model for furture values for each individual zip codes.

```python
zipcodes.to_csv('zipcode_data.csv')
```
