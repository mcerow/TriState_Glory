```python
df = pd.read_csv('zillow_data.csv')
df.head()
```

## Data Preprocessing

```python
def get_datetimes(df):
    return pd.to_datetime(df.columns.values[7:], format='%Y-%m')
```

```python
get_datetimes(df)
```

```python
df.shape
```

```python
df['ROI'] = (df['2018-04']/df['1996-04'])-1
```

## Filtering for Zipcodes

```python
tristate = df[df['State'].isin(['NY', 'NJ', 'CT'])]
tristate.shape
```

```python
tristate.isnull().sum()
```

```python
tristate.dropna(subset=['ROI'],inplace=True)
```

```python
tristate['AveragePrice'] = tristate.loc[:,'1996-04':'2018-04'].mean(axis=1)
```

```python
tristate.shape
```

```python
tristate.rename({'RegionName': 'Zipcode'}, axis='columns', inplace=True)
```

```python
tristate.sort_values('ROI',axis=0,ascending=False)[:5]
```

```python
new_york = tristate[tristate['State'] == 'NY']
new_jersey = tristate[tristate['State'] == 'NJ']
connecticut = tristate[tristate['State'] == 'CT']
```

```python
new_york[['Zipcode', 'City', 'State', 'Metro', 'CountyName', 'AveragePrice', 'ROI']].sort_values('ROI',axis=0,ascending=False)[:3]
```

```python
new_jersey[['Zipcode', 'City', 'State', 'Metro', 'CountyName', 'AveragePrice', 'ROI']].sort_values('ROI',axis=0,ascending=False)[:3]
```

```python
connecticut[['Zipcode', 'City', 'State', 'Metro', 'CountyName', 'AveragePrice', 'ROI']].sort_values('ROI',axis=0,ascending=False)[:3]
```
